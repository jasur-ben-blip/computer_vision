```python
import cv2
```

```python
from google.colab.patches import cv2_imshow

image = cv2.imread('/content/cat.jpg')
cv2_imshow(image)
```

```python
print("SHAPE:", image.shape) # (height, width, color chanels(RGB))
print("HEIGHT:", image.shape[0])
print("HEIGHT:", image.shape[0])
print("CHANELS:", image.shape[2]) # opencvda BGR => Blue, Green, Red
```

```python
(b, g, r) = image[100, 50]
print("Blue:", b, "Green:", g, "Red:", r)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(gray, (300, 300)) # (height, witdth)
cv2_imshow(resized)
```

```python
cropped = image[100:200, 300:500]
cv2_imshow(cropped)
```

```python
rotated = cv2.rotate(gray, cv2.ROTATE_180)
cv2_imshow(rotated)
```

```python
flipped = cv2.flip(gray, 1)
cv2_imshow(flipped)
```

```python
cv2.imwrite("flipped_cat.jpg", flipped)
```

```python
flipped_cat = cv2.imread("/content/flipped_cat.jpg")
cv2_imshow(flipped_cat)
```
